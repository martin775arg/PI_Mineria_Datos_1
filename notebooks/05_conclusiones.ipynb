# 1. Resumen del Proceso de Minería de Datos
A lo largo de este proyecto integrador, se ejecutó un pipeline completo de Ciencia de Datos sobre el comportamiento de la base de usuarios de la plataforma de streaming (8,160 registros iniciales):
* **Calidad de Datos:** Se identificaron y subsanaron inconsistencias críticas como formatos de fecha erróneos, categorías duplicadas en los planes de suscripción y valores atípicos físicos (*outliers* de hasta 99,999 minutos y edades imposibles).
* **Exploración (EDA):** Mediante análisis univariados y bivariados (tablas dinámicas y mapas de calor geográficos), descubrimos que la masa crítica de usuarios es adulta/joven y que el consumo en minutos es independiente del costo del plan contratado.
* **Modelado Avanzado (PCA):** La reducción de dimensionalidad demostró que el comportamiento del cliente está regido por tres perfiles ortogonales independientes de igual peso (~33.3% de varianza cada uno).

# ¿Existe fricción tecnológica asociada a la edad del cliente?
Sí. El Análisis de Componentes Principales (PC2) reveló una fuerte co-ocurrencia entre el volumen de tickets de soporte técnico abiertos y los rangos etarios más avanzados. Los usuarios de mayor edad presentan una tasa de incidentes significativamente más alta, marcando una clara fricción operativa en ese segmento.

### ¿Cómo interactúan el consumo, la edad y los reclamos?
No interactúan de manera lineal directa. Al ser dimensiones ortogonales en el espacio del PCA, coexisten perfiles muy marcados: el "Consumidor Maduro Fiel" (alta edad, alto consumo, cero reclamos), el "Usuario con Fricción" (alta edad, reclamos frecuentes, consumo variable) y el "Joven Maratoneador" (baja edad, consumo masivo, reclamos bajos).

## 3. Recomendaciones Estratégicas para la Plataforma

Con base en la evidencia analítica obtenida, se proponen tres líneas de acción urgentes para los equipos de toma de decisiones:

1. Rediseño de UX/UI enfocado en Accesibilidad (Equipo de Producto): Dado que el volumen de reclamos (PC2) está ligado a la edad, es mandatorio simplificar la interfaz de usuario, agrandar componentes visuales y automatizar tutoriales de navegación para mitigar la fricción técnica en adultos mayores.
2. Estrategia de Upselling de Valor Técnico (Equipo Comercial): Como el tiempo de visualización no cambia según el plan, las campañas para migrar usuarios del plan Básico al Premium no deben prometer "más contenido", sino centrarse exclusivamente en atributos de calidad técnica (4K, audio envolvente, descargas offline) y beneficios familiares (cuentas simultáneas).
3. Fidelización del Segmento Joven (Equipo de Marketing): El perfil PC3 ("Joven Maratoneador") es el motor de reproducción de la plataforma. Se recomienda diseñar sistemas de recomendación gamificados y campañas de retención específicas para este grupo, protegiendo su alta tasa de retención frente a competidores.
